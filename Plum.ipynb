# 1. Calculating The Efficiency Metrics:-

```python
df['First_reply_time_hours']=df['First reply time in minutes within business hours']/60
df['First_resolution_time_hours']=df['First resolution time in minutes within business hours']/60
df['Full_resolution_time_hours']=df['Full resolution time in minutes within business hours']/60
```

#  2. Performance of Different Groups :-

```python
group_stats = df.groupby('Group').agg({
    'First_reply_time_hours': 'mean',
    'First_resolution_time_hours': 'mean',
    'Full_resolution_time_hours': 'mean'
}).reset_index()
```

```python
total_first_reply_time=sum(group_stats['First_reply_time_hours'])
total_first_resolution_time=sum(group_stats['First_resolution_time_hours'])
total_full_resolution_time=sum(group_stats['Full_resolution_time_hours'])

labels=['First Reply Time','First Resolution Time','Full Resolution Time']
sizes=[total_first_reply_time,total_first_resolution_time,total_full_resolution_time]
colors=['Red','Skyblue','Orange']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels,colors=colors,autopct='%1.1f%%',startangle=140)
plt.axis('equal')
plt.title('Proportion of Total Time Spent on Each Metric')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax=sns.barplot(data=group_stats.melt(id_vars='Group'),x='Group',y='value',hue='variable')
plt.xticks()
plt.title('Group-wise Efficiency Metrics')
plt.xlabel('Group')
plt.ylabel('Hours')
plt.legend(title='Metrics')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for col in group_stats.columns[1:]:
    plt.plot(group_stats['Group'], group_stats[col], marker='o', label=col)
plt.xticks(rotation=45)
plt.title('Group-wise Efficiency Metrics')
plt.xlabel('Group')
plt.ylabel('Hours')
plt.legend()
plt.grid(True)
plt.show()
```

# 3. Ticket Types and Resolution Times:-

```python
ticket_types = df.groupby('Manual Tagging of Categories [list]').agg({
    'First_resolution_time_hours': 'mean'
}).reset_index().sort_values(by='First_resolution_time_hours',ascending=False)
```

```python
ticket_types
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=ticket_types.head(10),x='Manual Tagging of Categories [list]',y='First_resolution_time_hours')
plt.xticks(rotation=45)
plt.title('Ticket Types and Resolution Times')
plt.xlabel('Ticket Type')
plt.ylabel('Average Resolution Time (hours)')
plt.show()
```
